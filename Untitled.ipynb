### Modelagem Matemtica:
Queremos minimizar $(1-\alpha)E_o + \alpha E_N$, para $\alpha \in (0,1)$, onde $E_o$  referente  taxa de sobreposio dos retngulos e $E_N$ a quanto a soluo distorce a estrutura de vizinhana original dos dados.

$$
E_o = \frac{2}{n(n+1)} \sum_{i=1}^n\sum_{j=1}^n [O_{ij}(x,h)O_{ij}(y,v)]\\
$$

$$
\begin{split}
    O_{ij}(x,h) = \frac{1}{h_j^4} [h_j^2 - (x_i - x_j)^2]_+,\quad \textrm{se} \quad  & x_i \ge x_j\\
    O_{ij}(x,h) = \frac{1}{h_i^4} [h_i^2 - (x_i - x_j)^2]_+,\quad \textrm{se} \quad & x_i < x_j
\end{split}
$$

As definies so dadas a seguir.

```julia
x(x) = x  0 ? x : 0

O(i, j, x, h) = x[i]  x[j] ? x( h[j]^2 - (x[i] - x[j])^2 )^2 / h[j]^4 : 
                                x( h[i]^2 - (x[i] - x[j])^2 )^2 / h[i]^4
 
function E(xy, h, w)
    n = size(xy)[2]
    
    x = xy[1:2:end]
    y = xy[2:2:end]  
    retval = 0
    
    for i in 1:n     
        for j in (i + 1):n
            retval +=  O(i, j, x, h) *  O(i, j, y, w)
        end
    end 
    return 2 * retval / (n * (n + 1))
end
```
